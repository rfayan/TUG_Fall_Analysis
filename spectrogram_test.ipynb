```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
plt.interactive(True)
data = np.load('data_fusion.npy')
```

```python
dfft = np.fft.fft(data[0])
rfft = np.fft.rfft(data[0])
```

```python
f, t, Sxx = signal.spectrogram(dfft, fs=100)
plt.pcolormesh(t, f, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
plt.specgram(dfft, Fs=100)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.specgram(rfft, Fs=100)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.show()
```
