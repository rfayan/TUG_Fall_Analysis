```python
def print_summary(summary, title):
    ax = plt.subplot(221)
    plt.title('AUC_Score')
    summary['AUC_Score'].plot.bar(ax=ax)
    ax.grid(axis='y', alpha=0.5, linestyle='--')
    ax.get_xaxis().set_visible(False)

    ax = plt.subplot(222)
    plt.title('F1_Score')
    summary['F1_Score'].plot.bar(ax=ax)
    ax.grid(axis='y', alpha=0.5, linestyle='--')
    ax.get_xaxis().set_visible(False)

    ax = plt.subplot(223)
    plt.title('Average_Precision')
    summary['Average_Precision'].plot.bar(ax=ax)
    ax.grid(axis='y', alpha=0.5, linestyle='--')
    plt.xticks(rotation=0)

    ax = plt.subplot(224)
    plt.title('Balanced_Accuracy')
    summary['Balanced_Accuracy'].plot.bar(ax=ax)
    ax.grid(axis='y', alpha=0.5, linestyle='--')
    plt.xticks(rotation=0)

    plt.suptitle(title, fontsize=16)
    plt.show()
```

```python
method_summary = df_results.groupby(['Method']).agg({'AUC_Score':         [min, 'mean', max],
                                                     'F1_Score':          [min, 'mean', max],
                                                     'Average_Precision': [min, 'mean', max],
                                                     'Balanced_Accuracy': [min, 'mean', max]})
method_summary
```

```python
print_summary(method_summary, 'Classifiers Aggregate Results')
```

```python
signal_summary = df_results.groupby(['Signal']).agg({'AUC_Score':         [min, 'mean', max],
                                                     'F1_Score':          [min, 'mean', max],
                                                     'Average_Precision': [min, 'mean', max],
                                                     'Balanced_Accuracy': [min, 'mean', max]})
signal_summary
```

```python
print_summary(signal_summary, 'Signal Type Aggregate Results')
```

```python
features_summary = df_results.groupby(['Features']).agg({'AUC_Score':         [min, 'mean', max],
                                                         'F1_Score':          [min, 'mean', max],
                                                         'Average_Precision': [min, 'mean', max],
                                                         'Balanced_Accuracy': [min, 'mean', max]})
features_summary.iloc[0], features_summary.iloc[-1] = features_summary.iloc[-1].copy(), features_summary.iloc[0].copy()  # Swap first and last rows
features_summary = features_summary.rename(index={'1088': 'OriginalFeatures', 'OriginalFeatures': '1088'})               # for better graph visual
features_summary
```

```python
print_summary(features_summary, 'Features Aggregate Results')
```
